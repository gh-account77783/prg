```python
import matplotlib.pyplot as plt
import numpy as np

# --- Chart Data and Configuration ---

# Data extracted from the user-provided image (in ₹ 1000 Crores)
categories = ['Television', 'Room AC', 'Refrigerator', 'Washing Machine', 'Brown Goods']
years = ['FY 19-20', 'FY 24-25', 'FY 29-30F']
data = {
    'Television': [29, 38, 50],
    'Room AC': [17, 37, 91],
    'Refrigerator': [20, 35, 54],
    'Washing Machine': [17, 25, 35],
    'Brown Goods': [27, 43, 70],
}

# Convert data to a numpy array for easier manipulation
data_array = np.array(list(data.values()))

# Calculate the cumulative sum for stacking the bars
cumulative_data = np.cumsum(data_array, axis=0)
totals = cumulative_data[-1] # The last row of the cumulative sum is the total

# Define a color palette with shades of red and grey for a professional look
colors = ['#b22222', '#dc0e20', '#cd5c5c', '#f08080', '#ffb6c1']


# --- Creating the Plot ---

# Create a figure and a set of subplots with a white background
fig, ax = plt.subplots(figsize=(16, 8), facecolor='white')
ax.set_facecolor('white')

bar_width = 0.6

# Plot each category as a layer in the stacked bar chart
for i, category in enumerate(categories):
    # The 'bottom' parameter is used to stack the bars
    bottom_values = cumulative_data[i-1] if i > 0 else np.zeros(len(years))
    bars = ax.bar(years, data_array[i], bottom=bottom_values,
                  label=category, color=colors[i], width=bar_width)

    # Add data labels inside each segment
    for j, bar in enumerate(bars):
        height = data_array[i, j]
        # Only add label if the segment is large enough to hold it
        if height > 10:
             ax.text(bar.get_x() + bar.get_width() / 2,
                    bottom_values[j] + height / 2,
                    int(height),
                    ha='center',
                    va='center',
                    color='white',
                    fontsize=18,
                    fontweight='bold')


# --- Customizing the Plot's Appearance ---

# Add total labels on top of each bar
for i, total in enumerate(totals):
    ax.text(i, total + 5, f'{total}', ha='center', fontsize=22, fontweight='bold')

# Set the label for the Y-axis
ax.set_ylabel("Market Size (in ₹ 1,000 Crores)", fontsize=22, fontweight='bold')

# Increase the font size of the tick labels on both axes
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=14)

# Add a legend to the chart
ax.legend(fontsize=22, loc='upper left', frameon=False)

# Remove the top, right, and left spines for a cleaner look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('grey')

# Remove y-axis ticks
ax.tick_params(axis='y', length=0)
# Remove y-axis grid lines
ax.yaxis.grid(False)


# Ensure the layout is tight and nothing is clipped
plt.tight_layout()

# --- Display the Plot ---
plt.show()

# To save the chart to a file, you would use:
# plt.savefig('consumer_durables_breakdown.png', dpi=300, facecolor='white', bbox_inches='tight')
```
